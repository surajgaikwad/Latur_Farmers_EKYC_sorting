```python
df11 =[]
for i in range(len(df1)):
    df11.append(df1.iloc[[i],:].values.tolist())
```

```python
df11
```

```python
df22 =[]
for i in range(len(df2)):
    df22.append(df2.iloc[[i],:].values.tolist())
```

```python
df22
```

```python
df33 = []
for i in df11:
    if i not in df22:
        df33.append(i)
for j in df22:
    if j not in df11:
        df33.append(j)
```

```python
df33 = np.array(df33)
df33[0,:,:]
```
